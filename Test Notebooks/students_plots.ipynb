```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf
from tensorflow.keras.layers import Dense
from keras.callbacks import ModelCheckpoint
import numpy as np
#  Import and read the charity_data.csv.
import pandas as pd 
student_df  = pd.read_csv('Resources/students_adaptability.csv')
student_df.head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Change the font family and font name
plt.rcParams['font.family'] = 'sans-serif'  # Set the font family (e.g., sans-serif, serif)
plt.rcParams['font.sans-serif'] = ['Arial']  # Set the font name

# Assuming df is your DataFrame

# List of categorical variables to analyze
categorical_variables = ["Gender", "Education Level", "Institution Type", "Load-shedding", 
                         "Financial Condition", "Internet Type", "Self Lms", "Device"]

# Loop through each variable and create count plots
for var in categorical_variables:
    plt.figure(figsize=(10, 6))

    # Group the data by the variable and adaptability level and count the occurrences
    data_grouped = student_df.groupby([var, "Adaptivity Level"]).size().unstack().fillna(0)

    # Create a stacked bar chart
    data_grouped.plot(kind='bar', stacked=True)

    plt.title(f'Count Plot of {var} by Adaptability Level')
    plt.xlabel(var)
    plt.ylabel('Count')
    plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
    plt.legend(title="Adaptability Level")

    # Show the plot for the current variable
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming df is your DataFrame
adaptability_counts = student_df['Adaptivity Level'].value_counts()

# Create a bar plot
plt.bar(adaptability_counts.index, adaptability_counts)

# Customize the plot
plt.title('Distribution of Adaptability Levels')
plt.xlabel('Adaptability Level')
plt.ylabel('Number of Students')

# Show the plot
plt.show()
```
